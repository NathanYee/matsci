```python
import numpy as np
import seaborn as sns
import pandas as pd
import csv
%matplotlib inline
```

```python
%pylab inline
# pylab.rcParams['figure.figsize'] = (8, 6)
pylab.rcParams['figure.dpi'] = 200
```

```python
data = []
with open('Hardness Testing - Sheet2.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    for row in spamreader:
        sample = row[0]
        print(sample)
        tops = [[sample, float(x), 'Top'] for x in row[1:11]]
        bottoms = [[sample, float(x), 'Bottom'] for x in row[11:21]]
        data += tops + bottoms
```

```python
df = pd.DataFrame(data, columns=['Sample','Hardness','Side'])
```

```python
color0 = sns.color_palette('deep')[0]
color1 = sns.color_palette('deep')[1]
color2 = sns.color_palette('deep')[2]
```

```python
color0 = (*color0, .3)
```

```python
ax = sns.violinplot(x='Sample', y='Hardness', data=df, color=color0, width=.7, linewidth=.8)
```

```python
ax = sns.violinplot(x='Sample', y='Hardness', data=df,
                    inner=None, color=color0, width=.9, linewidth=.3)
ax = sns.swarmplot(x='Sample', y='Hardness', data=df, color=color1, size=3)
```

```python
sns.set_style('darkgrid')
```

```python
ax = sns.boxplot(x="Sample", y="Hardness", data=df, color=color0)
ax = sns.swarmplot(x='Sample', y='Hardness', data=df, size=3, color=color2)
ax.set(xlabel='Sample', ylabel='Hardness (HRC)')
```

```python
ax = sns.violinplot(x='Sample', y='Hardness', data=df, hue="Side",
                    palette='deep', split=True, scale="count", inner="quartile")
```

```python
ax = sns.violinplot(x='Sample', y='Hardness', data=df, inner=None)
```

```python

sns.set()

# Create a random dataset across several variables
rs = np.random.RandomState(0)
n, p = 40, 8
d = rs.normal(0, 2, (n, p))
d += np.log(np.arange(1, p + 1)) * -5 + 10

# Use cubehelix to get a custom sequential palette
pal = sns.cubehelix_palette(p, rot=-.5, dark=.3)

# Show each distribution with both violins and points
sns.violinplot(data=d, palette=pal, inner="points")
```
